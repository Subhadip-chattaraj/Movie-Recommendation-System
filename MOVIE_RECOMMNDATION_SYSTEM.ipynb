```
#title,cast,crew,overview,id,keywords,genres
final_movieframe=final_movieframe[['movie_id','keywords','title','crew','cast','overview','genres']]
final_movieframe.head()
```

```
final_movieframe.dropna(inplace=True)
```

```
final_movieframe.isnull().sum()
```

```
final_movieframe.genres
```

```
def convert(obj):
  anslist=[]
  for i in ast.literal_eval(obj):
    anslist.append(i['name'])
  return anslist
```

```
final_movieframe['genres']=final_movieframe['genres'].apply(convert)
```

```
final_movieframe['keywords']=final_movieframe['keywords'].apply(convert)
```

```
def convert_cast(obj):
  anslist=[]
  count=0
  for i in ast.literal_eval(obj):
    if count>2:
      break
    anslist.append(i['name'])
  return anslist
```

```
final_movieframe['cast']=final_movieframe['cast'].apply(convert_cast)
```

```
final_movieframe.head()
```

```
def convert_crew(obj):
  anslist=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      anslist.append(i['name'])
  return anslist
```

```
final_movieframe['crew']=final_movieframe['crew'].apply(convert_crew)
```

```
final_movieframe.head()
```

```
final_movieframe.head()
```

```
final_movieframe['overview']=final_movieframe['overview'].apply(lambda x:x.split())
```

```
final_movieframe.head(1)
```

```
final_movieframe['genres']=final_movieframe['genres'].apply(lambda x:[i.replace(" ","")for i in x])
final_movieframe['keywords']=final_movieframe['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
final_movieframe['cast']=final_movieframe['cast'].apply(lambda x:[i.replace(" ","")for i in x])
final_movieframe['crew']=final_movieframe['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```
final_movieframe.head(1)
```

```
final_movieframe['tags']=final_movieframe['overview']+final_movieframe['genres']+final_movieframe['keywords']+final_movieframe['cast']+final_movieframe['crew']
```

```
final_movieframe
```

```
recommender_df=final_movieframe[['movie_id','title','tags']]
```

```
recommender_df
```

```
import warnings
warnings.filterwarnings("ignore")
```

```
recommender_df['tags']=recommender_df['tags'].apply(lambda x:" ".join(x))
```

```
recommender_df['tags']=recommender_df['tags'].apply(lambda x:x.lower())
```

```
recommender_df
```

```
import nltk
```

```
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```
def stem(text):
  ans=[]
  for i in text.split():
    ans.append(ps.stem(i))

  return " ".join(ans)
```

```
recommender_df['tags']=recommender_df['tags'].apply(stem)
```

```
recommender_df
```

```
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```
vectors=cv.fit_transform(recommender_df['tags']).toarray()
```

```
vectors
```

```
vectors.shape
```

```
from sklearn.metrics.pairwise import cosine_similarity
```

```
similarity = cosine_similarity(vectors)
similarity[0]
```

```
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x :x[1])
```

```
def recommend(movie):
  movie_index = recommender_df[recommender_df['title']==movie].index[0]
  distances=similarity[movie_index]
  movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x :x[1])[1:6]
  for i in movie_list:
    print(recommender_df.iloc[i[0]].title)
```
